```python
import base64
import vertexai
from vertexai.generative_models import GenerativeModel, Part, FinishReason
import vertexai.preview.generative_models as generative_models

def generate(info,headlines,json1):
    vertexai.init(project="<project_name>", location="<location_here>")
    model = GenerativeModel("gemini-1.5-pro-001")
    prompt =f'''Carefully study the given information:{info}.
  For each headline in :{headlines}, check if it is relevant to or has an impact on any aspect on the operations of sricity plant. Provide reason for each of the headline (why and why not it is selected). 
  Consider all the regions the plant is connected too for picking the headlines. Pickout all relevant headlines.
  -----
  Output format:
  For all the picked headlines,
  1. headline: <selected headline here>
  2. Identify the location. 
  3. Determine the "type_of_disruption" based on the nature
  of the event (e.g., "weather", "people", "operational", "natural", etc.). 
  4. Determine the "impact", it can be "low", "medium", or "high" based on the headline context
  5. Specify the "severity" (should be a number from 1 to 5). 
  6. Describe the affected departments in "supply_chain_department_affected" 
  7. suggest a "recommended_course_of_action".  
  Few sample outputs:  {json1}
  Give output in this format only. follow the same format even if the headline is irrelevant
  '''
    #message_text = [{"role":"user","content":prompt}]
    responses = model.generate_content(
      [prompt],
      generation_config=generation_config,
      safety_settings=safety_settings,
      stream=True,
  )
    res = ''.join([i.text for i in responses])
    return res
       

generation_config = {
    "max_output_tokens": 8192,
    "temperature": 0.2,
    "top_p": 0.8,
}

safety_settings = {
    generative_models.HarmCategory.HARM_CATEGORY_HATE_SPEECH: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_HARASSMENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
}



op= generate(info, headlines[40:],json1)

print(op)

```
